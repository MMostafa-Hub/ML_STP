```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv("Position_Salaries.csv")
x = df["Level"].values.reshape(-1,1) # or df.iloc[:,1].values
y = df["Salary"].values.reshape(-1,1) # or df.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # splitting and randomizing the Data Set
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomiarize = PolynomialFeatures(degree=3) # fitting the data with a Hypothesis Function of 3rd Degree
x_train_poly = polynomiarize.fit_transform(x_train) # ploynomiarizing the x_train data Set
x_poly = polynomiarize.fit_transform(x) # ploynomiarizing X data set
```

X = 
\begin{bmatrix}
x^{0} &x^1&....& x^{n}
\end{bmatrix}
X_poly =
\begin{bmatrix}
(x^{(0)})^0 & (x^{(0)})^{1} & .... & (x^{(0)})^{n}\\
(x^{(1)})^0 & (x^{(1)})^{1}& .... & (x^{(1)})^{n}\\
. & . & .&.\\
. & . & .&.\\
(x^{(m)})^0 & (x^{(m)})^{1}& .... & (x^{(m)})^{n}\\
\end{bmatrix}


```python
from sklearn.linear_model import LinearRegression # takes multiVariable or Simple 
# NOTE: polynomial Regression is just a MultiVars Linear Regression but the X is polynomirized , so we polynormalize the x then we train it in a MultiVars Linear Regressiion model 

model = LinearRegression() # we must make an instance of LinearRegression once to save the Hypothesis Function model
model.fit(x_train_poly,y_train) # trainig and then saving the parameters of the Hypothesis function in model

H_XPoly = model.predict(x_poly) # predicted Y values of X   = H(x_poly)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y,H_XPoly) # calculating the score = 1 - cost of H(X_poly)
```
